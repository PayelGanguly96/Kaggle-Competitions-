```python
from os import path
from wordcloud import WordCloud
```

```python
d = path.dirname('C:/Users/payel/Desktop/model.txt')
```

```python
text = open(path.join(d, 'model.txt')).read() #Read the whole text.
```

```python
wordcloud = WordCloud().generate(text) #Generate a word cloud image
```

```python
import matplotlib.pyplot as plt    #Display the generated image
plt.imshow(wordcloud, interpolation='bilinear')  #imshow() displays an image on the axis 
plt.axis("off") 
```

```python
wordcloud = WordCloud(max_font_size=40).generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
